```python
import spacy
from nltk import ngrams
nlp=spacy.load('en_core_web_sm')
sentence=nlp("AI is the simulation of human intelligence by machine.It includes expect system NLP speech recognition and machine vision")
bigram=ngrams(sentence,2)
print("Bigrams:",list(bigram))
```

```python
import spacy
from nltk import ngrams
nlp=spacy.load('en_core_web_sm')
sentence1=nlp('AI is the simulation of human intelligence by machine.It includes expect system NLP speech recognition and machine vision')
trigram=ngrams(sentence1,3)
print("Trigrams:",list(trigram))
```
